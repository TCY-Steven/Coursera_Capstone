# Capstone Project
## Problem : Where would be recommended if someone want to open a restaurant in Queens ?


There are many neighborhoods in Queens, and there is a biggest train station in the Flushing. For the visitor in NY city, the traffic convenience would be a consideration point. So the location of the restaurant will affect the number of customer. By analysing the distance to train station and food related shop categories, we can get the information of the clustering of each neighborhood and we can know where is suitable for specific kind food shop or restaurant.
